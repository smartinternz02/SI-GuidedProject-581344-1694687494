```python
'''Inference from the above graph : from the above graph some are highly correlated (value >0.5) and
some are less correlated (less than <0.5) ex : here both the features total and speeding are highly
correlated because the value is greater than 0.61 which is greater than 0.5. if we take the features
total and ins_losses they are negativley correlated or we can say they are less correlated because
the value is -0.036 which is less than 0.5'''
```
